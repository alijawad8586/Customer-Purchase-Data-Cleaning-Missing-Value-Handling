```python
import pandas as pd
import numpy as np

data = {
    "age": [25, 30, np.nan, 40, 35, np.nan, 28, 50, np.nan, 45],
    "income": [50000, 60000, 55000, np.nan, 58000, 62000, np.nan, 900000, 57000, np.nan],
    "purchases": [3, 4, 2, 5, np.nan, 6, 3, 10, np.nan, 4],
  
}
df = pd.DataFrame(data)

print(df)
```

```python
df.to_csv("raw_customer_data.csv", index=False)
```

```python
print(df.isnull().sum())

missing_percentage = df.isnull().mean() * 100
print(missing_percentage)
```

```python
print(df.groupby(df["income"].isnull())["purchases"].mean())
```

```python
df["age"] = df["age"].fillna(df["age"].median())
```

```python
df["income"] = df["income"].fillna(df["income"].median())
```

```python
print(df.groupby(df["income"].isnull())["purchases"].mean())
```

```python
print(df["income"].isnull().value_counts())
```

```python
missing_report = pd.DataFrame({
    "Missing Count": df.isnull().sum(),
    "Missing Percentage": df.isnull().mean() * 100
})

print(missing_report)

missing_report.to_csv("missing_value_report.csv")
```
